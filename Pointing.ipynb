```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = ['/Users/schriste/Data/FOXSI-2-flight/LISS_pitch.txt', '/Users/schriste/Data/FOXSI-2-flight/LISS_yaw.txt']
```

```python
yaw = pd.read_csv(f[0], index_col=0, sep=" ", names=['yaw', 'no'])
pitch = pd.read_csv(f[1], index_col=0, sep=" ", names=['pitch', 'no'])
```

```python
yaw.drop('no', axis=1, inplace=True)
pitch.drop('no', axis=1, inplace=True)
```

```python
pointing = yaw
pointing = pointing.join([pitch])
```

```python
import datetime
```

```python
pointing['toffset'] = pointing.index
```

```python
launch_time = datetime.datetime(2014, 12, 11, 19, 11, 1)
```

```python
pointing['time'] = pd.DatetimeIndex(pd.to_datetime([datetime.timedelta(seconds = t)+launch_time for t in pointing.index]))
```

```python
pointing = pointing.set_index('time')
```

```python
pointing.plot(subplots=True);
```
